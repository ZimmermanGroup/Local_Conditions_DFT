Improved performance on enforcing constraint

```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
from sklearn.linear_model import Ridge


# Data
x = np.array([2, 3, 4, 0, 0, 0, 0, -12, -11, -10])
X = np.reshape(x, (len(x), 1))
y = np.array([4, 5, 7, 0, 0, 0, 0, 0, 0.5, 0.7])

# Unconstrained Ridge Regression
clf = Ridge(alpha=5.0)
clf.fit(X, y)

# Calculate MAE for unconstrained ridge regression
y_pred_unconstrained = clf.predict(X)
mae_unconstrained = np.mean(np.abs(y - y_pred_unconstrained))


# Plot unconstrained line
min_x, max_x = np.min(x), np.max(x)
x_range = np.linspace(min_x, max_x)
X_range = np.reshape(x_range, (len(x_range), 1))
y_range = clf.predict(X_range)
plt.plot(x_range, y_range, label='Unconstrained')

# Constrained Ridge Regression
m = cp.Variable(1)
c = cp.Variable(1)
objective = cp.Minimize(cp.sum_squares(m * x + c - y) + 5.0 * cp.sum_squares(m))
constraints = [m * x + c >= 0]
prob = cp.Problem(objective, constraints)
result = prob.solve()
mv, cv = m.value[0], c.value[0]

# Calculate MAE for constrained ridge regression
y_pred_constrained = mv * x + cv
mae_constrained = np.mean(np.abs(y - y_pred_constrained))


# Plot constrained line
y_constrained = mv * x_range + cv
plt.plot(x_range, y_constrained, label='Constrained')

# Plot data points
plt.plot(x, y, 'o')

# Display MAEs as text on the plot
plt.text(0.1, 0.75, f'MAE Unconstrained: {mae_unconstrained:.2f}', transform=plt.gca().transAxes, fontsize = 12)
plt.text(0.1, 0.70, f'MAE Constrained: {mae_constrained:.2f}', transform=plt.gca().transAxes, fontsize = 12)

plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)

# Labels and legend

plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.legend(fontsize = 12)

plt.show()
```
